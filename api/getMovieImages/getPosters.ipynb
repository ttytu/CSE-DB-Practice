```python
import pandas as pd

filename = "movie.csv"
df = pd.read_csv(filename)
df
```

```python
df = df[['movie title']]
df
```

```python
df['year'] = df['movie title'].str[-5:-1]
df['movie title'] = df['movie title'].str[:-6]
df
```

```python
base_url = "https://api.themoviedb.org/3/search/movie?query="

df['query'] = df['movie title'].str.replace(' ', '%20')
df['query'] = base_url + '?' + df['query']

df['movie title'] = df['movie title'].str.replace(r'\s*\([^()]*\)', '')
df
```
